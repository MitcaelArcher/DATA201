```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Question** - What is the demographics of the reported encounters of patients on diabetes meds?

```python
df=pd.read_csv("https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv")
df=df.set_index('encounter_id')
#This is reading the CSV and changing my index to the encounter ID
```

```python
df
```

```python
df.shape
#Shows there are 49 Columns and 101,766 Rows
```

```python
df.columns
#Detailing the 49 Columns
```

```python
df = df[df['diabetesMed'] == 'Yes']
#This is selecting the patients that are on diabetes meds
```

```python
df=df.drop(['admission_source_id','time_in_hospital','num_procedures','discharge_disposition_id','admission_type_id','num_medications','payer_code', 'weight','medical_specialty', 'num_lab_procedures', 'number_outpatient','number_emergency', 'number_inpatient', 'diag_1', 'diag_2', 'diag_3','number_diagnoses', 'max_glu_serum', 'A1Cresult', 'metformin','repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride','acetohexamide', 'glipizide', 'glyburide', 'tolbutamide','pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone','tolazamide', 'examide', 'citoglipton', 'insulin','glyburide-metformin', 'glipizide-metformin','glimepiride-pioglitazone', 'metformin-rosiglitazone','metformin-pioglitazone', 'change', 'diabetesMed', 'readmitted'], axis=1)
#Removing all columns that I do not need to complete my analysis
```

```python
df = df.rename(columns = {'patient_nbr':'Patient Number', 'race':'Race', 'gender':'Gender', 'age':'Age',}, inplace = False)
#This is changing the name of some columns
```

```python
df.columns
#Displaying the columns after removing the columns I did not want
```

```python
df['Gender'].value_counts()
#This is showing the count of encounters by gender
```

```python
sns.set(font_scale=1.4)
df['Gender'].value_counts().plot(kind='pie', figsize=(10, 7), rot=0)
plt.title("Count of Encounters by Gender", y=1.02);
#This is a pie graph showing the count of encounters by gender
```

```python
df['Age'].value_counts()
#This is showing the count of encounters by age
```

```python
sns.set(font_scale=1.4)
df['Age'].value_counts().plot(kind='barh', figsize=(10, 7), rot=0)
plt.xlabel("Age", labelpad=14)
plt.ylabel("Count of Encounters", labelpad=14)
plt.title("Count of Encounters by Age", y=1.02);
#This is a horizontal bar graph showing the count of encounters by age
```

```python
df['Race'].value_counts()
#This is showing the count of encounters by race
```

```python
sns.set(font_scale=1.4)
df['Race'].value_counts().plot(kind='bar', figsize=(12, 7), rot=0)
plt.xlabel("Race", labelpad=14)
plt.ylabel("Count of Encounters", labelpad=14)
plt.title("Count of Encounters by Race", y=1.02);
#This is a bar graph showing the count of people by race
```

Conclusion: Since the initial question for my project was, "What is the demographics of the reported encounters of patients on diabetes meds"? I was able to identify that the majority of the patients on diabetes meds are caucasian, between 70 & 80 years old, and are female.
